#Modxo Build script

```python
from pathlib import Path
import os

zipp = Path('.').resolve() / 'modxo.zip'
picosdk = Path('.').resolve() / "pico-sdk"
modxo = Path('.').resolve() / "modxo"
build = Path('.').resolve() / "modxodev"
with open("modxo.sh","w") as script:
  script.writelines(
      [
          f'export PICO_SDK_PATH="{str(picosdk)}"\n',
          f"cd {str(build)}\n",
          f"cmake {str(modxo)}\n",
          "make\n",
          f"zip -r {str(zipp)} {str(build)}\n"
      ]
  )
```
